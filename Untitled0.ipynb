```python
# Step 0: Libraries import
import pandas as pd
import numpy as np
from google.colab import files

# Step 1: Upload CSV file
print("Upload your CSV file:")
uploaded = files.upload()  # ye popup kholta hai, CSV select karo

# Print uploaded file info
for fn in uploaded.keys():
    print('User uploaded file "{name}" with length {length} bytes'.format(
        name=fn, length=len(uploaded[fn])))

# Step 2: Load dataset
# Replace filename with uploaded file name
filename = list(uploaded.keys())[0]  # automatically pick uploaded file
data = pd.read_csv(filename)

# Step 3: Explore dataset
print("\nFirst 5 rows of dataset:")
display(data.head())

print("\nDataset Info:")
print(data.info())  # columns, datatypes, non-null counts

print("\nStatistical Summary:")
print(data.describe())  # basic stats like mean, std, min, max

print("\nMissing Values in Each Column:")
print(data.isnull().sum())  # check missing values
```
